### Goal 

 * Plot subway entrance and exits on a map with the subway lines overlayed.
 * Don't use AI (ChatGTP, Claude, Copilot, etc.)

 Start with following [this tutorial](https://towardsdatascience.com/geopandas-101-plot-any-data-with-a-latitude-and-longitude-on-a-map-98e01944b972) for Chicago data.

```python
import pandas as pd
import matplotlib.pyplot as plt
import descartes
import geopandas as gpd
from shapely.geometry import Point, Polygon

%matplotlib inline
```

```python
# read in the data for the subway lines
subway_lines = gpd.read_file('data/Subway Lines/geo_export_a43cfd92-dc9c-4159-91f6-934c58291474.shp')
```

```python
# plot the subway lines 
fig, ax = plt.subplots(figsize=(15,15))
subway_lines.plot(ax=ax, alpha = 0.4, color = 'gray')
ax.set_axis_off();
```

```python
# read in the subway stops list
df = pd.read_csv('data/MTA_Subway_Entrances_and_Exits__2024_20250118.csv')
crs = 'epsg:4269' # coordinate reference system
df.head()
```

```python
geometry=gpd.points_from_xy(df['Entrance Longitude'], df['Entrance Latitude'])
geo_df = gpd.GeoDataFrame(df,
                          crs=crs, 
                          geometry=geometry)
geo_df.head()
```

```python
fig, ax = plt.subplots(figsize=(15,15))
subway_lines.plot(ax=ax, alpha=0.4, color='grey')
geo_df.plot(ax=ax,
            markersize=20,
            color='blue',
            marker='o',
            label='Subway Entrances/Exits')
plt.legend(prop={'size':15})
ax.set_axis_off();
```
